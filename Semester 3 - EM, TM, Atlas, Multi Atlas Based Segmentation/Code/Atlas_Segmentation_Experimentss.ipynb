## Functions for Creating the Atlas

```python
# -*- coding: utf-8 -*-
"""
Created on Thu Nov 14 17:23:55 2019


@authors: Zohaib & Isaac (MAIA_3, 3rd Semester, Girona)
"""

import os
import time
import cv2 as cv
import numpy as np
import nibabel as nib
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
from scipy.stats import moment
from scipy.stats import multivariate_normal
from sklearn.datasets import make_spd_matrix
from glob import glob

def getImgs(folder):
    '''
    This function explores the folders inside './results' path, where different cases exist.
    '''
    root_folder = os.getcwd()
    data_folder = os.path.join(root_folder, folder)
    folders = glob(data_folder + "/*")
    header,affine = get_header(folders[0])
    img = [img for folder in folders for img in load_images_from_folder(folder)]
    return img,header,affine

def load_images_from_folder(folder):
    images = []
    test=os.listdir(folder)
    test.sort()
    #print(test)
    for filename in test:
        if (filename.endswith(".nii") or filename.endswith(".nii.gz")):
            #print(filename)
            img = nib.load(os.path.join(folder, filename))
            img = img.get_fdata()

            if img is not None:
                if filename.endswith("g.nii"):
                    continue
                else:
                    images.append(img)
    return images

def get_header(folder):
    for filename in os.listdir(folder):
        if (filename.endswith(".nii") or filename.endswith(".nii.gz")):
            img = nib.load(os.path.join(folder, filename))
            head = img.header
            aff = img.affine
    return head,aff

img,header,affine = getImgs('registered_images')
gt_img,_,_ = getImgs('registered_labels')

```

## Tissue Model (Normalized Probabilities)


```python
scale_factor = 1000
bins_no = 1000
bins = np.arange(0,bins_no)
print(len(bins))
img = load_images_from_folder('training/training-images')
gt_img = load_images_from_folder('training/training-labels')

for i in range(0,len(gt_img)):
  img[i] = img[i].astype(int)
  normalized_img =((img[i])- np.min(img[i]))/(np.max(img[i])-np.min(img[i]))
  normalized_img = normalized_img * scale_factor
  img[i] = normalized_img.astype(int)


  if(i == 0):    
    hist_CSF ,_= np.histogram(img[i][gt_img[i] == 1],bins)
    hist_GM ,_= np.histogram(img[i][gt_img[i] == 3],bins)
    hist_WM ,_= np.histogram(img[i][gt_img[i] == 2],bins)
  else:
    hist_CSF_temp ,_= np.histogram(img[i][gt_img[i] == 1],bins)
    hist_GM_temp ,_= np.histogram(img[i][gt_img[i] == 3],bins)
    hist_WM_temp ,_= np.histogram(img[i][gt_img[i] == 2],bins)
    hist_CSF += hist_CSF_temp
    hist_GM += hist_GM_temp
    hist_WM += hist_WM_temp

total = hist_CSF + hist_GM + hist_WM
hist_CSF = np.divide(hist_CSF ,total)
hist_GM = np.divide(hist_GM ,total)
hist_WM = np.divide(hist_WM ,total)


plt.grid(axis='y', alpha=0.75)
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})

plt.plot(bins[:-1], hist_CSF, color="#3F5D7D", label='CSF')


plt.plot(bins[:-1], hist_WM, color='c', label='Grey Matter')

plt.plot(bins[:-1], hist_GM, color="k", label='White Matter')

plt.gca().set(title='Tissue Model', ylabel='Tissue Model Probabilities')
plt.legend()
plt.savefig("tissue_model_uc.png")
plt.show()
```

## Tissue Model Cleaned

```python
def cleanTM(t1, t2, t3):
    where_are_NaNs = np.isnan(t3)
    t3[where_are_NaNs] = 0
    h = (np.where(t3 == max(t3))[0][0])
    t3[h:] = 1
    t2[h:] = 0
    t1[h:] = 0
    return t1, t2, t3,h

hist_CSF_n,hist_WM_n,hist_GM_n,h= cleanTM(hist_CSF,hist_GM,hist_WM)

plt.grid(axis='y', alpha=0.75)
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})

plt.plot(bins[:-1], hist_CSF_n[:], color="#3F5D7D", label='CSF')


plt.plot(bins[:-1], hist_GM_n[:], color='c', label='Grey Matter')

plt.plot(bins[:-1], hist_WM_n[:], color="k", label='White Matter')

plt.gca().set(title='Cleaned Tissue Model', ylabel='Probability Corresponding to each intensity')
plt.legend()
plt.savefig("tissue_model_c.png")
plt.show()
```

```python
def load_images_num_from_folder(folder,num,flag = 0):
    images = []
    test = os.listdir(folder)
    test.sort()
    filename = test[num]
    #print(test)
    if (filename.endswith(".nii") or filename.endswith(".nii.gz")):
        if (flag == 1):
          i=0
          #print("Processing Test Image: " + filename)
        img = nib.load(os.path.join(folder, filename))
        img = img.get_fdata()

    return img


def load_image_from_folder_num(folder,num):
    root_folder = os.getcwd()
    data_folder = os.path.join(root_folder, folder)
    folders = glob(data_folder + "/*")
    #print(folders)
    folders.sort()
    folder = folders[num]
    #print(folder)
    images = []
    for filename in os.listdir(folder):
        if (filename.endswith(".nii") or filename.endswith(".nii.gz")):
            #print(filename)
            img = nib.load(os.path.join(folder, filename))
            img = img.get_fdata()
    return img


def dice(seg, gt):
    dice_val = (2*np.sum(gt*seg))/(np.sum(gt) + np.sum(seg))
    return dice_val  

def calculate_dice_scores(seg, gt,n="temp"):

    labelCSF = 1
    labelGM = 3
    labelWM = 2
        
    CSF = (seg == labelCSF) * 1
    CSFgt = (gt == labelCSF) * 1

    GM = (seg == labelGM) * 1
    GMgt = (gt == labelGM) * 1

    WM = (seg == labelWM) * 1
    WMgt = (gt == labelWM) * 1
    
    plt.figure() 
    plt.subplot(1,3,1)
    plt.imshow(CSF[:,:,130])

    plt.subplot(1,3,2)
    plt.imshow(WM[:,:,130])

    plt.subplot(1,3,3)
    plt.imshow(GM[:,:,130])
    plt.savefig(n)
    plt.show()
    CSFdice = dice(CSF, CSFgt)
    GMdice = dice(GM, GMgt)
    WMdice = dice(WM, WMgt)
    
    return [CSFdice, GMdice, WMdice]
```

## Function for Predicting Probablistic Atlas

```python

def get_atlas_segmentation(image_no):
  i = image_no
  testing_mask = load_images_num_from_folder('testing/testing-mask/',i)
  testing_labels = load_images_num_from_folder('testing/testing-labels/',i)
  testing_images = load_images_num_from_folder('testing/testing-images/',i,1)


  WM = load_image_from_folder_num('predictions/WM/',i)
  GM = load_image_from_folder_num('predictions/GM/',i)
  CSF = load_image_from_folder_num('predictions/CSF/',i)


  WM = WM * (testing_mask > 0)
  GM = GM * (testing_mask >0)
  CSF = CSF * (testing_mask >0)

  CSFlabel = 1
  GMlabel = 2
  WMlabel = 3

  GM_final = GMlabel * (np.greater(GM ,CSF) * np.greater(GM , WM))
  WM_final = WMlabel * (np.greater(WM , CSF) * np.greater(WM , GM))
  CSF_final = CSFlabel * (np.greater(CSF , WM) * np.greater(CSF , GM))
  

  segmentation = GM_final + WM_final + CSF_final
  test = calculate_dice_scores(segmentation, testing_labels)
  return test,segmentation,CSF,WM,GM
```

```python
def init(gt, img, flag = 'kmeans',atlas =0):
  '''
  Initialization for the EM. We can have a k-means , or atlas/TM initialization
  . The variable atlas is used to pass both TM and atlas initializations
  '''

  imgSk = []
  if(flag == 'kmeans'):
    labels_store = []
    kmeansCent = []
    label_mean = []
    label_cov = []
    label_class_ratio = []

    lenGt = len(gt)
    for i in range(lenGt):
        labels_store = np.argwhere(gt[i] > 0)
        t1_idx_i, t1_idx_j, t1_idx_k = zip(*labels_store)
        f = []
        for m in range(int(len(img)/lenGt)):
            reshaped = img[2*i+m][t1_idx_i,t1_idx_j,t1_idx_k]
            class_cal = img[2*i+m][t1_idx_i,t1_idx_j,t1_idx_k]
            f = (np.expand_dims(reshaped,-1))
            if m == 0:
                input_concat = f
            else:
                input_concat = np.concatenate((input_concat, f), axis=1)
        
        kmeans = KMeans(n_clusters=3, random_state=0).fit(input_concat)
        imgSk.append(kmeans.labels_)
        print("Processing Image No: ", i+1, '/', lenGt)
        labels_store = np.squeeze(np.argwhere(kmeans.labels_ == 0))
        label0_mean = (np.mean(input_concat[labels_store],axis=0))
        label0_cov = (np.cov(input_concat[labels_store,:], rowvar = False))
        ratio_label0 = (labels_store.shape[0]/class_cal.shape[0])

        labels_store = np.squeeze(np.argwhere(kmeans.labels_ == 1))
        label1_mean = (np.mean(input_concat[labels_store],axis=0))
        label1_cov = (np.cov(input_concat[labels_store,:], rowvar = False))
        ratio_label1 = (labels_store.shape[0]/class_cal.shape[0])

        labels_store = np.squeeze(np.argwhere(kmeans.labels_ == 2))
        label2_mean = (np.mean(input_concat[labels_store],axis=0))
        label2_cov = (np.cov(input_concat[labels_store,:], rowvar = False))
        ratio_label2 = (labels_store.shape[0]/class_cal.shape[0])

        mean_val = [label0_mean,label1_mean,label2_mean]
        cov_val = [label0_cov,label1_cov,label2_cov]
        ratio_val = [ratio_label0,ratio_label1,ratio_label2]

        label_mean.append(mean_val)
        label_cov.append(cov_val)
        label_class_ratio.append(ratio_val)
        kmeansCent.append(kmeans.cluster_centers_)
  if flag == 'atlas':
    labels_store = []
    kmeansCent = []
    label_mean = []
    label_cov = []
    label_class_ratio = []

    lenGt = len(gt)
    for i in range(lenGt):
      labels_store = np.argwhere(gt[i] > 0)
      t1_idx_i, t1_idx_j, t1_idx_k = zip(*labels_store)
      f = []
      for m in range(int(len(img)/lenGt)):
        reshaped = img[2*i+m][t1_idx_i,t1_idx_j,t1_idx_k]
        class_cal = img[2*i+m][t1_idx_i,t1_idx_j,t1_idx_k]
        f = (np.expand_dims(reshaped,-1))
        if m == 0:
            input_concat = f
        else:
            input_concat = np.concatenate((input_concat, f), axis=1)
      prediction = atlas[t1_idx_i,t1_idx_j,t1_idx_k]
      imgSk.append(prediction)
      #print("Processing Image No: ", i+1, '/', lenGt)

      labels_store = np.squeeze(np.argwhere(prediction == 1))
      label0_mean = (np.mean(input_concat[labels_store],axis=0))
      label0_cov = (np.cov(input_concat[labels_store,:], rowvar = False))
      ratio_label0 = (labels_store.shape[0]/class_cal.shape[0])

      labels_store = np.squeeze(np.argwhere(prediction == 2))
      label1_mean = (np.mean(input_concat[labels_store],axis=0))
      label1_cov = (np.cov(input_concat[labels_store,:], rowvar = False))
      ratio_label1 = (labels_store.shape[0]/class_cal.shape[0])

      labels_store = np.squeeze(np.argwhere(prediction == 3))
      label2_mean = (np.mean(input_concat[labels_store],axis=0))
      label2_cov = (np.cov(input_concat[labels_store,:], rowvar = False))
      ratio_label2 = (labels_store.shape[0]/class_cal.shape[0])

      mean_val = [label0_mean,label1_mean,label2_mean]
      cov_val = [label0_cov,label1_cov,label2_cov]
      ratio_val = [ratio_label0,ratio_label1,ratio_label2]

      label_mean.append(mean_val)
      label_cov.append(cov_val)
      label_class_ratio.append(ratio_val)
        
  return label_mean,label_cov,label_class_ratio,kmeansCent, imgSk

def EM_dice_cal(imgS, testing_labels):
  test1 =[dice(imgS[0] == 1,testing_labels==1),dice(imgS[0] ==2,testing_labels==1),dice(imgS[0] ==3,testing_labels==1)]
  test2 =[dice(imgS[0] == 1,testing_labels==2),dice(imgS[0]==2,testing_labels==2),dice(imgS[0]==3,testing_labels==2)]
  test3 = [dice(imgS[0] == 1,testing_labels==3),dice(imgS[0]==2,testing_labels==3),dice(imgS[0]==3,testing_labels==3)]



  label2_no = np.argmax(test2) 
  label3_no = np.argmax(test3)
  label1_no = 3-label2_no - label3_no
  label2dice = np.max(test2)
  label3dice = np.max(test3)
  label1dice = test1[3-label2_no - label3_no]

  print(label1dice,label2dice,label3dice)

  plt.figure() 
  plt.subplot(1,3,1)
  plt.imshow((imgS[0][:,:,130] == (label1_no+1)))

  plt.subplot(1,3,2)
  plt.imshow(imgS[0][:,:,130] == (label2_no+1))

  plt.subplot(1,3,3)
  plt.imshow(imgS[0][:,:,130] == (label3_no+1))
  plt.show()


def match1Lab(gt, img, imgA):
    ''' Returns the correct labelling of tissues
    gt : Ground Truth array
    img: Segmented images array
    imgA: Grayscale image
    '''
    new_img = np.zeros(img.shape)

    label1 = np.sum(imgA * (img ==1))/np.sum((img ==1))
    label2 = np.sum(imgA * (img ==2))/np.sum((img ==2))
    label3 = np.sum(imgA * (img ==3))/np.sum((img ==3))

    maxlabel = np.argmax([label1,label2,label3]) 
    minlabel = np.argmin([label1,label2,label3]) 
    otherlabel = 6 - maxlabel - minlabel

    new_img = new_img + ((img == maxlabel) * 3)
    new_img = new_img + ((img == minlabel) * 1)
    new_img = new_img + ((img == otherlabel) * 2)

    return new_img


def matchLab(gt, img, imgA):
    '''
    gt : Ground Truth arrays
    img: Segmented images arrays
    imgA: Grayscale images
    '''
    lenGt = len(gt)
    mods = int(len(imgA)/lenGt)
    labs = len(np.unique(gt[0]))
    for i in range(lenGt):
        mg = []
        mi = []
        for lab in range(1,labs):
            mg.append(np.mean(moment((gt[i]==lab) * imgA[i*mods], moment = 2)))
            mi.append(np.mean(moment((img[i]==lab) * imgA[i*mods], moment = 2)))
        minG = mg.index(min(mg))
        minI = mi.index(min(mi))
        maxG = mg.index(max(mg))
        maxI = mi.index(max(mi))
        if minI != minG or maxI != maxG:
            img[i][img[i] == 0] = 4
            img[i][img[i] == minI+1] = minG + 5
            img[i][img[i] == maxI+1] = maxG + 5
            img[i][img[i] == (3 - minI - maxI+1)] = (3 - minG - maxG) + 5
            img[i] = img[i] - 4
        
    return img

def Ex_Maximize(label_mean,label_cov,label_class_ratio,gtA, imgM, CSF,GM,WM,itera = 200, conv = 0.01,flag_conv = 0,include_atlas = 0 ):
  ''' E.M. algorithm
  '''


  tol = 1e-8
  output = []
  nonzero_i= []
  nonzero_j = []
  nonzero_k = []
  iters = []
  max_iter = itera
  min_change = conv
  current_change =100

  prev_change = 0
  lenGt = len(gtA)
  mods = int(len(imgM)/lenGt)

  for i in range(lenGt):
    mean_dist = label_mean[i]
    cov_dist = label_cov[i]
    class_dist = label_class_ratio[i]
    prob_dist= [0,0,0]
    weight_dist = [0,0,0]

    labels_store = np.argwhere(gtA[i] > 0)
    t1_idx_i, t1_idx_j, t1_idx_k = zip(*labels_store)
    nonzero_i.append(t1_idx_i)
    nonzero_j.append(t1_idx_j)
    nonzero_k.append(t1_idx_k)

    for m in range(mods):
      reshaped = imgM[mods*i+m][t1_idx_i,t1_idx_j,t1_idx_k]

      f = (np.expand_dims(reshaped,-1))
      if m == 0:
          input_concat = f
      else:
          input_concat = np.concatenate((input_concat, f), axis=1)


    n_iter = 0
    absChange = itera + 1
    t0 = time.time()
    while (n_iter < max_iter and (absChange > min_change) or n_iter < 10 ):
      prev_change = current_change
      n_iter = n_iter + 1
      prob_dist[0] = multivariate_normal.pdf(input_concat, mean=mean_dist[0], cov=cov_dist[0]) 
      prob_dist[1] = multivariate_normal.pdf(input_concat, mean=mean_dist[1], cov=cov_dist[1]) 
      prob_dist[2] = multivariate_normal.pdf(input_concat, mean=mean_dist[2], cov=cov_dist[2]) 

      if (include_atlas != 0):
        prob_dist[0] = prob_dist[0] * CSF[t1_idx_i,t1_idx_j,t1_idx_k]
        prob_dist[1] = prob_dist[1] * GM[t1_idx_i,t1_idx_j,t1_idx_k]
        prob_dist[2] = prob_dist[2] * WM[t1_idx_i,t1_idx_j,t1_idx_k]

      weight_dist[0] = np.nan_to_num(prob_dist[0])
      weight_dist[1] = np.nan_to_num(prob_dist[1])
      weight_dist[2] = np.nan_to_num(prob_dist[2])

      total = weight_dist[0] + weight_dist[1] + weight_dist[2]

      weight_dist[0] = np.divide(weight_dist[0],total)
      weight_dist[1] = np.divide(weight_dist[1],total)
      weight_dist[2] = np.divide(weight_dist[2],total)

      #weight_dist[0] = (prob_dist[0] * class_dist[0]) /((prob_dist[0] * class_dist[0]) + (prob_dist[1] * class_dist[1]) + (prob_dist[2] * class_dist[2]))
      #weight_dist[1] = (prob_dist[1] * class_dist[1]) / ((prob_dist[0] * class_dist[0]) + (prob_dist[1] * class_dist[1]) + (prob_dist[2] * class_dist[2]))
      #weight_dist[2] = (prob_dist[2] * class_dist[2]) / ((prob_dist[0] * class_dist[0]) + (prob_dist[1] * class_dist[1]) + (prob_dist[2] * class_dist[2]))
      
      weight_dist[0] = np.nan_to_num(weight_dist[0])
      weight_dist[1] = np.nan_to_num(weight_dist[1])
      weight_dist[2] = np.nan_to_num(weight_dist[2])   
      
      weight_vector = np.vstack((weight_dist[0],weight_dist[1],weight_dist[2]))

      prediction = np.argmax(weight_vector, axis=0)

      #class_dist[0] = np.mean(weight_dist[0])
      #class_dist[1] = np.mean(weight_dist[1])
      #class_dist[2] = np.mean(weight_dist[2])


      mean_dist[0] = np.dot(np.expand_dims(weight_vector[ 0,:],0) , input_concat)/ (np.sum(weight_dist[0])+tol)
      mean_dist[1] = np.dot(np.expand_dims(weight_vector[ 1,:],0) , input_concat)/ (np.sum(weight_dist[1])+tol)
      mean_dist[2] = np.dot(np.expand_dims(weight_vector[ 2,:],0) , input_concat)/ (np.sum(weight_dist[2])+tol)

      cov_dist[0] =  np.dot((np.expand_dims(weight_vector[ 0,:],1) * (input_concat - mean_dist[0])).transpose() , (input_concat - mean_dist[0])) /(np.sum(weight_dist[0])+tol)
      cov_dist[1] =  np.dot((np.expand_dims(weight_vector[ 1,:],1) * (input_concat - mean_dist[1])).transpose() , (input_concat - mean_dist[1])) /(np.sum(weight_dist[1])+tol)
      cov_dist[2] =  np.dot((np.expand_dims(weight_vector[ 2,:],1) * (input_concat - mean_dist[2])).transpose() , (input_concat - mean_dist[2])) /(np.sum(weight_dist[2])+tol)


      mean_dist[0] = np.squeeze(mean_dist[0].transpose())
      mean_dist[1] = np.squeeze(mean_dist[1].transpose())
      mean_dist[2] = np.squeeze(mean_dist[2].transpose())
      
      #print(n_iter)

      if (flag_conv == 0):
        absChange = 1
      else:
        current_change = np.sum(np.log10((class_dist[0] * weight_dist[0]) + (class_dist[1] * weight_dist[1]) + (class_dist[2] * weight_dist[2])))
        absChange = abs(current_change - prev_change)


    iters.append(n_iter)
    #print('this case took ', n_iter, 'iterations and ', round(time.time()-t0,1), 'seconds')
    output.append(prediction)
  return nonzero_i,nonzero_j,nonzero_k,output, iters
```

```python
def histogram_prob(img, hist,val_len=1000):
    output = np.zeros(img.shape, dtype='float')
    for i in range(0,val_len-1):
        output[img == i] = hist[i]
    return output
```
